```python
import os
import tarfile
from six.moves import urllib
import pandas as pd
from pandas.plotting import scatter_matrix
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

## **FETCH AND LOAD DATA**

```python
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = os.path.join("datasets","housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url = HOUSING_URL, housing_path = HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path,"housing.tgz")
    urllib.request.urlretrieve(housing_url,tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path = housing_path)
    housing_tgz.close()
```

## **CREATE A TEST SET** 

```python
## Write a test set segregation function
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
```

It is important that we have representive data on our hands so that we arent victims of sampling error. Suppose a population consists of 51% males and 49% females, even if we have a small sample, we try to maintain the same ratio. This technique is *stratified sampling*.

Imagine if we are told that the median income is a key predictor. We want to be in a confident position that the data we have is representative of the population in a healthy way. This means we need sufficient number of instances/observations for each stratum, else the importance of the stratum might be biased. Thus we should not have too many strata and each strata must be large enough.

Lets stratify our median income varaible/predictor, by bucketing it.

```python
housing["income_cat"] = np.ceil(housing["median_income"]/1.5)
housing["income_cat"].hist()
```

```python
housing["income_cat"].where(housing["income_cat"]<5, 5.0,inplace = True)
housing["income_cat"].hist()
```

```python
# Stratified Sampling
split = StratifiedShuffleSplit(n_splits=1,test_size= 0.2, random_state=42)
```

```python
for train_index,test_index in split.split(housing,housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
len(strat_train_set) 
```

```python
strat_train_set["income_cat"].value_counts()/len(strat_train_set)
```

```python
strat_test_set["income_cat"].value_counts()/len(strat_test_set)
```

As can be seen, stratified sampling can be used to generate a test set thats a more appropriate representation of the entire dataset; as compared to randomly sampling out a test set.  

```python
# reset the dataframe to original state, for the sake of experimenting
for ds in (strat_train_set, strat_test_set):
    ds.drop("income_cat",axis=1,inplace=True)
```

### Data Cleaning - Missing Values - Imputing 

There are various ways to deal with missing data, namely - 
- Dropping the predictor entirely
- Dropping the districts (rows) with missing values
- Filling in the missing values

Needless to say, the first 2 methods can lead to a loss of some leverageable data.

```python
# option 1
housing.drop("total_bedrooms",axis=1)
# option 2
housing.dropna(subset = ["total_bedrooms"])
#option 3
median_tot_bedrooms = housing.total_bedrooms.median()
housing.total_bedrooms.fillna(median_tot_bedrooms,inplace = True)
```
